```python
bike = pd.read_csv('bike.csv')
bike
```

```python
bike['생년'].replace('\\N', pd.NA, inplace=True)
bike['성별'].replace('\\N', pd.NA, inplace=True)

NaN_bike = bike[bike[['생년', '성별']].isna().any(axis=1)]

# 생년과 성별의 NaN 값을 가진 행을 기존 데이터프레임에서 제거
bike4 = bike.dropna(subset=['생년', '성별'])
```

```python
bike4['대여일자'] = pd.to_datetime(bike4['대여일자'])
bike4['대여일자'] = bike4['대여일자'].dt.strftime('%Y-%m-%d-%H')
bike4
```

```python
# 성별 컬럼에서 'M'과 'F' 외의 값은 '기타'로 처리
bike4['성별'] = bike4['성별'].replace({'M': '남성', 'F': '여성'}).fillna('기타')

# 생년을 9년 단위로 그룹화
bins = list(range(1900, 2025, 10))  # 1900년부터 2020년까지 9년 단위로 그룹화
labels = [f'{year}~{year+9}' for year in bins[:-1]]  # 라벨링
bike4['생년대'] = pd.cut(bike4['생년'].astype(float), bins=bins, labels=labels, right=False)

bike4 = bike4[bike4['생년'].astype(float).between(1920, 2020)]

bike4.head()
```

```python
gender_grouped = bike4.groupby(['대여일자', '성별']).size().unstack(fill_value=0).reset_index()
gender_grouped
```

```python
gender_grouped = gender_grouped.rename(columns={'F': '여성', 'M': '남성'})
```

```python
gender_grouped['남성'] = gender_grouped['남성'] + gender_grouped['f']
gender_grouped['여성'] = gender_grouped['여성'] + gender_grouped['m']
gender_grouped
```

```python
bike_sum = bike4.groupby('대여일자').size().reset_index(name='대여')
bike_sum
```

```python
user_type_grouped = bike4.groupby(['대여일자', '이용자종류']).size().unstack(fill_value=0).reset_index()
user_type_grouped
```

```python
birth_grouped = bike4.groupby(['대여일자', '생년대']).size().unstack(fill_value=0).reset_index()
birth_grouped
```

```python
merged_df = pd.merge(gender_grouped, user_type_grouped, on='대여일자', how='outer')
merged_df = pd.merge(merged_df, birth_grouped, on='대여일자', how='outer')
merged_df = pd.merge(merged_df, bike_sum, on='대여일자', how='outer')
```
